```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0,5,11)
a
```

```python
b = a**2
b
```

```python
c = np.array([2,4,6,8,7])
c
```

```python
d = c**2
d
```

```python
plt.plot(a,b)
plt.title('Salary',fontsize=25,c='orange')
plt.xlabel('A',fontsize=18,c='red')
plt.ylabel('B',fontsize=18,c='red')
plt.show()
```

```python
plt.scatter(a,b)
plt.title('Salary',fontsize=25,c='blue')
plt.xlabel('A',fontsize=18)
plt.ylabel('B',fontsize=18)
plt.show()
```

```python
plt.plot(c)
plt.title('salary',fontsize=20)
plt.xlabel('A',fontsize=18)
plt.ylabel('B',fontsize=18)
plt.show()
```

```python
plt.bar(a,b)
plt.title('Salary',fontsize=20)
plt.xlabel('A',fontsize=18)
plt.ylabel('B',fontsize=18)
plt.show()
```

```python
plt.bar(c,d)
plt.title('Salary',fontsize=20)
plt.xlabel('A',fontsize=18)
plt.ylabel('B',fontsize=18)
plt.show()
```

# subplots

```python
plt.subplot(1,3,1)
plt.plot(a,b)
plt.subplot(1,3,2)
plt.plot(b,a)
plt.subplot(1,3,3)
plt.plot(a,a)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])   #left,bottom,width,height
axes.plot(a,b)
axes.set_title('SparrowAI')
axes.set_xlabel('A')
axes.set_ylabel('B')
plt.show()
```
