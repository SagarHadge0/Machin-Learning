```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

# Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:,1].values
```

```python
x
```

```python
y = dataset.iloc[:,2].values
```

```python
y
```

```python
from sklearn.svm import SVR
```

```python
regressor = SVR(kernel ='sigmoid')
```

```python
regressor.fit(x.reshape(-1,1),y)
```

```python
regressor.score(x.reshape(-1,1),y)
```

```python
def predict_salary(position_level):
    position_level_standardized = (np.array([[position_level]]))
    
    # Predict the salary using the SVR model
    predicted_salary_standardized = regressor.predict(position_level_standardized)
    
    # Inverse transform the predicted salary to the original scale
    predicted_salary =(predicted_salary_standardized.reshape(-1, 1))
    
    return predicted_salary[0][0]
```

```python
new_position_level = 6
predicted_salary = predict_salary(new_position_level)
print(f'Predicted salary for position level {new_position_level}{predicted_salary:.2f}')
```

# Random_forest

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
dataset1 = pd.read_csv('Position_Salaries.csv')
```

```python
dataset1
```

```python
x = dataset1.iloc[:,1].values.reshape(-1,1)
```

```python
x
```

```python
y = dataset1.iloc[:,2].values
```

```python
y
```

```python
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
```

```python
regressor
```

```python
regressor.fit(x,y)
```

```python
def predict_salary(position_level):
    predicted_salary = regressor.predict(np.array([[position_level]]))
    return predicted_salary[0]
```

```python
new_position_level = 1
```

```python
predicted_salary = predict_salary(new_position_level)
```

```python
print(f'Predicted salary for position level {new_position_level}{predicted_salary:.2f}')
```
