# Imputing Missing Values

```python
test = np.array([[2,2],[3,4],[4,5],[np.nan,np.nan]])
```

```python
test
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
test_1 = imp.fit_transform(test)
```
