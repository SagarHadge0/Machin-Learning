# Agglomarative

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set_style('darkgrid')
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
x = iris.data
```

```python
x
```

```python
x.shape
```

```python
x=x[0:50,:]
```

```python
x.shape
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
model = AgglomerativeClustering(distance_threshold=1.5,n_clusters=None)
```

```python
model.fit(x)
```

```python
model.fit_predict(x)
```

```python
model.n_clusters_
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
z = hierarchy.linkage(model.children_,'ward')
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(z)
```

```python
identified_cluster = model.fit_predict(x)
```

```python
identified_cluster
```

```python
data = pd.DataFrame(x)
```

```python
data
```

```python
data['cluster'] = identified_cluster
```
